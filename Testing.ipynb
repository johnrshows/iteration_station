```python
#import dependencies
import csv
import pandas as pd
```

```python
#reading in traffic data from csv
Csvpath = "US_Accidents_Dec19.csv"
traffic_data = pd.read_csv(Csvpath)
traffic_data.head()
```

```python
#Finding all column names in traffic dataframe
traffic_data.columns
```

```python
#updating dataframe to keep only columns we need
traffic_data_reorg = traffic_data[["Severity", "Description", "Start_Time", "Start_Lat","Start_Lng","Street", "Zipcode", "City","County","State","Visibility(mi)"]]
traffic_data_reorg.head(30)
```

```python
#Narrowing down dataframe to only show Atlanta, Georgia accidents from 2019
date = '2019-01-01 00:00:00'
atl_data = traffic_data_reorg.loc[(traffic_data_reorg["State"]=="GA")
                                  &(traffic_data_reorg["City"]=="Atlanta")
                                  &(traffic_data_reorg["Start_Time"]>=date),:]
atl_data
```

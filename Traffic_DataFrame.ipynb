```python
#import dependencies
import csv
import pandas as pd
```

```python
#reading in traffic data from csv
Csvpath = "US_Accidents_Dec19.csv"
traffic_data = pd.read_csv(Csvpath)
traffic_data.head()
```

```python
#Finding all column names in traffic dataframe
traffic_data.columns
```

```python
#updating dataframe to keep only columns we need
traffic_data_reorg = traffic_data[["Severity", "Description", "Start_Time", "Start_Lat","Start_Lng", "Zipcode", "City","County","State"]]
traffic_data_reorg.head()
```

```python
#Narrowing down dataframe to only show Atlanta, Georgia accidents from 2019
date = '2019-01-01 00:00:00'
atl_data = traffic_data_reorg.loc[(traffic_data_reorg["State"]=="GA")
                                  &(traffic_data_reorg["City"]=="Atlanta")
                                  &(traffic_data_reorg["Start_Time"]>=date),:]
atl_data
```

```python
#finding zipcodes with highest number of accidents
zip_cts = atl_data["Zipcode"].value_counts()
zip_cts = pd.DataFrame(zip_cts)
zip_cts.head(10)
```

```python
#narrowing down dataframe to only show zipcodes with highest number of accidents
narrowed_atl_df = atl_data.loc[(atl_data["Zipcode"]=='30312')
                               |(atl_data["Zipcode"]=='30339')
                               |(atl_data["Zipcode"]=='30331')
                               |(atl_data["Zipcode"]=='30315')
                               |(atl_data["Zipcode"]=='30309')
                               |(atl_data["Zipcode"]=='30340')
                               |(atl_data["Zipcode"]=='30308')
                               |(atl_data["Zipcode"]=='30324')
                               |(atl_data["Zipcode"]=='30342')
                               |(atl_data["Zipcode"]=='30318'),:]
narrowed_atl_df
```

```python
#converting Start Time to datetime dtype and adding new column with corrected date format
narrowed_atl_df.dtypes
narrowed_atl_df['Start_Time']= pd.to_datetime(narrowed_atl_df['Start_Time']) 
narrowed_atl_df['Date'] = narrowed_atl_df['Start_Time'].dt.strftime('%m/%d/%Y')
narrowed_atl_df
```

```python
narrowed_atl_df.groupby('Date').count()
```

```python
#create empty dataframe with dates
atl_accident_df = pd.DataFrame(index=pd.date_range("01/01/2019", periods=365))
atl_accident_df.index = atl_accident_df.index.strftime('%m/%d/%Y')
atl_accident_df
```

```python
# for loop needed to add all accident data in zipcodes of interest to dataframe

#create list of zipcodes
zipcodes = ["30312","30339","30331", "30315","30309","30340","30308","30324","30342","30318"]


#loop through each zipcode in dataframe to get number of accidents each day
#add new column to empty dataframe with data of each zipcode
for zipcode in zipcodes:
    zip_df = narrowed_atl_df.loc[narrowed_atl_df["Zipcode"] == zipcode, :]
    num_accidents = zip_df.groupby("Date").count()
    num_accidents = num_accidents["State"]
    atl_accident_df[zipcode] = num_accidents
    
#display dataframe   
atl_accident_df
```

```python
#Name index
#replace Nan with '0'
atl_accident_df.index.name = 'Date'
atl_accident_df.fillna(0)
```

```python
atl_accident_df.to_csv('Atl_Traffic_data.csv',index = True, header = True)
```
